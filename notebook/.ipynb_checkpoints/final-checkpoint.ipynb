# MY OWN MODEL FEATURES

```python
import spacy
import pandas as pd

# Function to create strings with overlap, lemmatize, and store the results
def create_strings_with_overlap(paragraph, overlap=3, words_per_string=8):
    words = paragraph.split()

    result_strings = []
    for i in range(0, len(words) - words_per_string + 1, words_per_string - overlap):
        result_strings.append(' '.join(words[i:i+words_per_string]))

    # Include the remaining words at the end
    remaining_words = words[-(words_per_string-overlap):]
    if remaining_words:
        result_strings.append(' '.join(remaining_words))

    return result_strings

nlp = spacy.load("en_core_web_sm")

# Function to lemmatize text and remove stop words
def lemmatize_text(text):
    doc = nlp(text)
    lemmatized_text = " ".join([token.lemma_ for token in doc if (
        not token.is_stop or token.text.lower() in ['no', 'not']) and not token.is_punct])
    return lemmatized_text

# Function to create strings with overlap, lemmatize, and store the results
def process_jokes(row):
    joke_number = row['Joke Number']
    text = row['Text']

    # Replace "Q." with "question" and "A." with "answer"
    text = text.replace('Q.', 'question').replace('A.', 'answer').replace('q.', 'question').replace('a.', 'answer').replace('q:', 'question').replace('a:', 'answer')

    # Remove stop words
    text = ' '.join([word for word in text.split() if word.lower() not in nlp.Defaults.stop_words])

    # Create strings with overlap
    strings_with_overlap = create_strings_with_overlap(text)

    # Lemmatize each string
    lemmatized_strings = [lemmatize_text(string) for string in strings_with_overlap]

    # Create a new DataFrame for the results
    result_data = {'Index': list(range(1, len(lemmatized_strings) + 1)),
                   'Joke Number': [joke_number] * len(lemmatized_strings),
                   'Lemmatized Text': lemmatized_strings}

    result_df = pd.DataFrame(result_data)
    return result_df

# Apply the processing function to each row in jokes_df
processed_dfs = jokes_df.apply(process_jokes, axis=1)

# Concatenate the results into a single DataFrame
result_df = pd.concat(processed_dfs.tolist(), ignore_index=True)

# Save the DataFrame to a CSV file
result_df.to_csv('lemmatized_jokes.csv', index=False)

# Display the new DataFrame
print(result_df)
```
